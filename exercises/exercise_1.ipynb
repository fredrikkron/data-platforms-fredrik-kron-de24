```python
file_path = "data/riket2023_åk9_np.xlsx"

df_math = pd.read_excel(file_path, sheet_name="Matematik")
df_math
```

```python
df_cleaned2 = df_math.tail(4).rename( {
        "Unnamed: 0": "Plats",
        "Unnamed: 1": "Huvudman",
        "Unnamed: 2": "Totalt (A-F)",
        "Unnamed: 3": "Flickor (A-F)",
        "Unnamed: 4": "Pojkar (A-F)",
        "Unnamed: 5": "Totalt (A-E)",
        "Unnamed: 6": "Flickor (A-E)",
        "Unnamed: 7": "Pojkar (A-E)",
        "Unnamed: 8": "Totalt (poäng)",
        "Unnamed: 9": "Flickor (Poäng)",
        "Unnamed: 10": "Pojkar (poäng)"
    },
    axis=1,
)
df_cleaned2
```

```python
df_cleaned2["Totalt (A-F)"] = df_cleaned2["Totalt (A-F)"].astype(int)
```

```python
df_cleaned2
```

```python
df_math_total = df_cleaned2.query("Huvudman == 'Samtliga'")
df_math_total["Totalt (A-F)"]
```

```python
df_percent = df_math_total["Totalt (A-F)"] * (1 - (df_math_total["Totalt (A-E)"] / 100))
df_percent = df_percent.astype(int)
df_percent
```
