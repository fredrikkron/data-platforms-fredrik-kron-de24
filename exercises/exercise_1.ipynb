# 1.2 Olympic games data

### a)

```python
df1 = pd.read_csv("data/athlete_events.csv")
df1.head()
```

### b)

```python
import duckdb

df_olympic = duckdb.query("SELECT * FROM df1").df()
```

### c)

```python
df1.info()
```

### d)

```python
query1 = """
WITH ranked_athletes AS (
    SELECT
        Name,
        MAX(Age) AS Age,
        Sport
    FROM
        df_olympic
    WHERE
        Age IS NOT NULL
    GROUP BY
        Name, Sport
)
SELECT
    DISTINCT ON (Name) Name,
    Age,
    Sport
FROM
    ranked_athletes
ORDER BY
    Age DESC
LIMIT 10
"""

print(query1)
```

```python
ten_oldest_athletes = duckdb.query(query1).df()
ten_oldest_athletes
```

### e)

```python
query2 = """
WITH ranked_athletes AS (
    SELECT
        Name,
        MIN(Age) AS Age,
        Sport
    FROM
        df_olympic
    WHERE
        Age IS NOT NULL
    GROUP BY
        Name, Sport
)
SELECT
    DISTINCT ON (Name) Name,
    Age,
    Sport
FROM
    ranked_athletes
ORDER BY
    Age
LIMIT 10
"""
```

```python
ten_youngest_athletes = duckdb.query(query2).df()
ten_youngest_athletes
```

### f)

```python
query3 = """
SELECT
    MEDIAN(Age) AS median_age,
    Sport
FROM
    df_olympic
WHERE
    Age IS NOT NULL
GROUP BY
    Sport
ORDER BY median_age DESC
LIMIT 5
    """
```

```python
five_sports_with_highest_age = duckdb.query(query3).df()
five_sports_with_highest_age
```

### g)

```python
query4 = """
SELECT
    MEDIAN(Age) as median_age,
    Sport
FROM
    df_olympic
GROUP BY
    Sport
ORDER BY
    median_age
LIMIT 5
"""
```

```python
five_sport_with_lowest_age = duckdb.query(query4).df()
five_sport_with_lowest_age
```

### h)

```python
query5 = """
SELECT 
  Team,
  COUNT(Medal) as Gold_medals
FROM df_olympic
WHERE 
  Medal = 'Gold'
GROUP BY
  Team
ORDER BY
  Gold_Medals DESC
LIMIT 10
"""
```

```python
top_10_gold_medals = duckdb.query(query5).df()
top_10_gold_medals.index = range(1, len(top_10_gold_medals) + 1)
top_10_gold_medals
```

### i)

```python
query6 = """
SELECT 
  Team,
  COUNT(Medal) as total_medals
FROM df_olympic
GROUP BY
  Team
ORDER BY
  total_medals DESC
LIMIT 10
"""
```

```python
top_10_all_medals = duckdb.query(query6).df()
top_10_all_medals.index = range(1, len(top_10_all_medals) + 1)
top_10_all_medals
```

### j)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group by Year and Sex, and count unique athletes (based on ID)
gender_count = df_olympic.groupby(['Year', 'Sex'])['ID'].nunique().reset_index()
# Rename columns for clarity
gender_count.rename(columns={'ID': 'AthleteCount'}, inplace=True)
gender_count

# Plot using Seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=gender_count,
    x='Year',
    y='AthleteCount',
    hue='Sex',
    marker='o'
)

# Customize the plot
plt.title('Number of Male and Female Athletes Over Time', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Athletes', fontsize=12)
plt.legend(title='Gender', loc='upper left')
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```
