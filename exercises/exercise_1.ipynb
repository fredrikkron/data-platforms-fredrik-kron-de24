```python
df1 = pd.read_csv("data/athlete_events.csv")
df1.head()
```

### b)

```python
import duckdb

df_olympic = duckdb.query("SELECT * FROM df1").df()
```

### c)

```python
df1.info()
```

### d)

```python
query1 = """
WITH ranked_athletes AS (
    SELECT
        Name,
        MAX(Age) AS Age,
        Sport
    FROM
        df_olympic
    WHERE
        Age IS NOT NULL
    GROUP BY
        Name, Sport
)
SELECT
    DISTINCT ON (Name) Name,
    Age,
    Sport
FROM
    ranked_athletes
ORDER BY
    Age DESC
LIMIT 10
"""

print(query1)
```

```python
ten_oldest_athletes = duckdb.query(query1).df()
ten_oldest_athletes
```

### e)

```python
query2 = """
WITH ranked_athletes AS (
    SELECT
        Name,
        MIN(Age) AS Age,
        Sport
    FROM
        df_olympic
    WHERE
        Age IS NOT NULL
    GROUP BY
        Name, Sport
)
SELECT
    DISTINCT ON (Name) Name,
    Age,
    Sport
FROM
    ranked_athletes
ORDER BY
    Age
LIMIT 10
"""
```

```python
ten_youngest_athletes = duckdb.query(query2).df()
ten_youngest_athletes
```

### f)

```python
query3 = """
SELECT
    MEDIAN(Age) AS median_age,
    Sport
FROM
    df_olympic
WHERE
    Age IS NOT NULL
GROUP BY
    Sport
ORDER BY median_age DESC
LIMIT 5
    """
```

```python
five_sports_with_highest_age = duckdb.query(query3).df()
five_sports_with_highest_age
```

### g)

```python
query4 = """
SELECT
    MEDIAN(Age) as median_age,
    Sport
FROM
    df_olympic
GROUP BY
    Sport
ORDER BY
    median_age
LIMIT 5
"""
```

```python
five_sport_with_lowest_age = duckdb.query(query4).df()
five_sport_with_lowest_age
```
